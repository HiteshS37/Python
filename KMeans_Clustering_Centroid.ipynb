```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = load_wine()
wine_data = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
scaler = StandardScaler()
wine_data_scaled = scaler.fit_transform(wine_data)
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
kmeans.fit(wine_data_scaled)
```

```python
wine_data['Cluster'] = kmeans.labels_
```

```python
clusters_centers = scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(x = wine_data_scaled[:, 0], y = wine_data_scaled[:, 1], hue=wine_data['Cluster'], palette = 'viridis', legend = 'full')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c ='red',label = 'Centroids')
plt.title('K-Means Clustering of Wine Dataset')
plt.xlabel("Feature 1 (Srandardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()
```
