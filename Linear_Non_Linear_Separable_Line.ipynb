```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC

for sep, title in [(2, "Linear"), (0.5, "Non-Linear")]:
  X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, random_state=42, class_sep=sep)
  clf = SVC(kernel='linear').fit(X, y)
  plt.scatter(X[:, 0], X[:,1], c=y, cmap='coolwarm')
  w = clf.coef_[0]
  a = -w[0] / w[1]
  xx = np.linspace(X[:, 0].min(), X[:, 0].max())
  yy = a * xx - clf.intercept_[0] / w[1]
  plt.plot(xx, yy, 'k-')
  plt.title(f"{title} Seperability")
  plt.show()
```
