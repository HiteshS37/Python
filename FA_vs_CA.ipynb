FACTOR ANALYSIS

```python
import pandas as pd
from sklearn.decomposition import FactorAnalysis
import matplotlib.pyplot as plt
```

```python
# Load dataset (example: Iris datatset simplicity)
from sklearn.datasets import load_iris
data = load_iris()
X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
# Perform Factor Analysis
factor = FactorAnalysis (n_components=2, random_state=0)
X_factor = factor.fit_transform(X)
```

```python
# Plot the factors
plt.scatter(X_factor[:, 0], X_factor[:, 1])
plt.xlabel('Factor 1')
plt.ylabel('Factor 2')
plt.title('Factor Analysis')
plt.show()
```

CLUSTER ANALYSIS

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset (example: Iris datatset simplicity)
from sklearn.datasets import load_iris
data = load_iris()
X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
# Perform KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
X['Cluster'] = kmeans.fit_predict(X)
```

```python
# Plot the Clusters
sns.scatterplot(x=X.iloc[:, 0 ], y=X.iloc[:,1], hue=X['Cluster'], palette='viridis')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title('K-means Clustering')
plt.show()
```
